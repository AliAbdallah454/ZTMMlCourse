## Use matplotlib OO Method most of the cases

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# pyplot method
over_50.plot(kind="scatter", x="age", y="chol", c="target")
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax)
ax.set_xlim([45, 70])
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
ax.set(xlabel="Age", ylabel="Chol", title="Relation between Chol and Age")

# Adding a legend
ax.legend(*scatter.legend_elements(), title="target");
ax.axhline(over_50["chol"].mean(), linestyle="dashed")
```

```python
over_50["target"].values
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
fig, ax = plt.subplots()

scat = ax.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target']
                     )
ax.set(xlabel="Age", ylabel="Thalach", title="Relationship between Age and Thalach")

ax.legend(*scat.legend_elements(), title="target")
ax.axhline(over_50['thalach'].mean())
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10))
# Add data to ax0

scat = ax0.scatter(x=over_50['age'],
                   y=over_50['chol'],
                   c=over_50['target'])
ax0.set(title="Hear disease and chol levels", xlabel='age', ylabel='chol')

ax0.axhline(over_50['chol'].mean(), linestyle="--")
ax0.legend(*scat.legend_elements(), title="target")

# Adding to ax1

scat2 = ax1.scatter(x=over_50['age'],
                    y=over_50['thalach'],
                    c=over_50['target'])
ax1.set(title="thalach vs age", xlabel="Age", ylabel="thalach")
ax1.legend(*scat2.legend_elements(), title="target")
ax1.axhline(over_50['thalach'].mean(), linestyle='--')

fig.suptitle("Heart Disease Analysis", fontsize=20, fontweight="bold")
```
